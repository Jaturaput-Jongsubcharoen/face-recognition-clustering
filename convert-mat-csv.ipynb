```python
import scipy.io as sio

mat = sio.loadmat("umist_cropped.mat")

data = mat["facedat"]
print("Data shape:", data.shape)

dirnames = mat["dirnames"]
print("Dirnames shape:", dirnames.shape)
print(dirnames)
```

```python
df = pd.DataFrame(data.T)
df.to_csv("umist_cropped.csv", index=False)
```

```python
# Flatten dirnames into Python list
names = [d[0] for d in dirnames[0]]

labels = []
for i, name in enumerate(names):
    # Example: number of images in each folder
    num_images = data.shape[1] // len(names)
    labels += [i] * num_images

# Add to dataframe
df["label"] = labels
```
